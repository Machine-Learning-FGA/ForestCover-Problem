## Visualizando os dados

```python
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
from sklearn.manifold import TSNE
import numpy as np 

pca = TSNE(n_components=2)

print('Iniciando redução de feature')

vis = pca.fit_transform(data_2.iloc[:, :-1],data_2.iloc[:,-1])

print('PCA complete!')

data_array = np.array(data_2.iloc[:])

colors = {1:'red', 2: 'orange', 3: 'yellow', 4:'green', 5:'blue', 6:'indigo', 7:'violet'}

print(vis.shape[0])

for i in range(vis.shape[0]):
    print(i, end='\r')
    plt.scatter(vis[i,0], vis[i,1], c=colors[data_array[i,-1]])

print('Plot Complete')

plt.show()
```

Pelo plot dos dados podemos concluir que as classes estão inseparaveis utilizando todas as features.

## Feature Selection

### Using Tree-Based Selection

```python
from sklearn.feature_selection import SelectFromModel
from sklearn.tree import DecisionTreeClassifier
```

```python
select = SelectFromModel(DecisionTreeClassifier())

data_selection = select.fit_transform(data.iloc[:,:-1], data.iloc[:,-1])
```

```python
print(data_selection)
```

```python
data_base = pd.DataFrame(data_selection
                        )
```
